```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# --- Problem Tanımı ---
# Amacımız: "Cat" ve "Dog" resimlerini sınıflandırmak
# Veri kümesi: 'PetImages' klasöründe kedi ve köpek fotoğrafları
# Hedef: Verilen bir resmin kedi mi köpek mi olduğunu tahmin etmek

DATADIR = "PetImages"
CATEGORIES = ["Cat", "Dog"]
IMG_SIZE = 100
MAX_IMAGES = 500  # Veri miktarı sınırı

data, labels = [], []

# --- Veri Hazırlama & Görüntü Ön İşleme ---
for label, category in enumerate(CATEGORIES):
    folder = os.path.join(DATADIR, category)
    count = 0
    for img_name in os.listdir(folder):
        if count >= MAX_IMAGES:
            break
        try:
            img_path = os.path.join(folder, img_name)
            img = cv2.imread(img_path)
            if img is None:
                continue
            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))  # Resize
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Renk kanallarını düzenle
            data.append(img)
            labels.append(label)
            count += 1
        except:
            continue

data = np.array(data) / 255.0  # Normalize et
labels = np.array(labels)

# Veri bölme
X_train_val, X_test, y_train_val, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.1, random_state=42)

# Veri artırma
datagen = ImageDataGenerator(rotation_range=20,
                             zoom_range=0.15,
                             horizontal_flip=True,
                             width_shift_range=0.15,
                             height_shift_range=0.15)
datagen.fit(X_train)

# --- Modelleme (CNN) ---
cnn_model = Sequential([
    Input(shape=(IMG_SIZE, IMG_SIZE, 3)),
    Conv2D(32, (3,3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    
    Conv2D(64, (3,3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    
    Conv2D(128, (3,3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    
    Flatten(),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.4),
    Dense(2, activation='softmax')
])

cnn_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005), 
                  loss='sparse_categorical_crossentropy', 
                  metrics=['accuracy'])

cnn_model.fit(datagen.flow(X_train, y_train, batch_size=32),
              epochs=25,
              validation_data=(X_val, y_val))

# --- Performans Değerlendirme ---
cnn_preds = np.argmax(cnn_model.predict(X_test), axis=1)
accuracy = accuracy_score(y_test, cnn_preds)
print(f"CNN Test Doğruluk Oranı: {accuracy*100:.2f}%")

cm = confusion_matrix(y_test, cnn_preds)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("CNN Confusion Matrix")
plt.xlabel("Tahmin")
plt.ylabel("Gerçek")
plt.show()

# ROC-AUC
y_test_bin = tf.keras.utils.to_categorical(y_test)
cnn_probs = cnn_model.predict(X_test)
auc = roc_auc_score(y_test_bin, cnn_probs)
print(f"CNN ROC-AUC: {auc:.4f}")

fpr, tpr, _ = roc_curve(y_test_bin[:,1], cnn_probs[:,1])
plt.plot(fpr, tpr, label=f"CNN ROC curve (area = {auc:.4f})")
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

# Sonuç görselleştirme (5 örnek)
for i in range(5):
    plt.imshow(X_test[i])
    plt.title(f"Gerçek: {CATEGORIES[y_test[i]]} - Tahmin: {CATEGORIES[cnn_preds[i]]}")
    plt.axis('off')
    plt.show()
```
